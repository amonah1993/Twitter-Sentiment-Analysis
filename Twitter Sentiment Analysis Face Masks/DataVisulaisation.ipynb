```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objs as go
import plotly.figure_factory as ff
```

```python
covidMaskP2=pd.read_csv('CovidMaskPeriod2PredictedBERTSentiment.csv')
covidMaskP1=pd.read_csv('CovidMaskPeriod1PredictedBERTSentiment.csv')
```

```python
# plotting First and Second periods tweets sentiments by month
# Source: https://stackoverflow.com/questions/56181597/matplotlib-line-chart-with-count Retrieved July 2022
```

```python
#plotting First period tweets sentiments by month

covidMaskP1['created_at'] = pd.Categorical(covidMaskP1['created_at'],
                                   categories=['Jun', 'Jul', 'Aug', 'Sep'],
                                   ordered=True)
pivotPlot = covidMaskP1.pivot_table(index = 'created_at', 
                          columns = 'Sentiment', 
                          fill_value = 0, 
                          aggfunc={'created_at':len})['created_at']
pivotPlot.plot.line(title='First period tweets sentiments by month')
plt.tight_layout()
plt.xlabel(" Month ")
plt.ylabel(" Sentiment Frequency ")
plt.show()
```

```python
#plotting Second period tweets sentiments by month

covidMaskP2['created_at'] = pd.Categorical(covidMaskP2['created_at'],
                                   categories=['Jun', 'Jul', 'Aug', 'Sep'],
                                   ordered=True)

pivotPlot = covidMaskP2.pivot_table(index = 'created_at', 
                          columns = 'Sentiment', 
                          fill_value = 0, 
                          aggfunc={'created_at':len})['created_at']
pivotPlot.plot.line(title='Second period tweets sentiments by month')
plt.tight_layout()
plt.xlabel(" Month ")
plt.ylabel(" Sentiment Frequency ")
plt.show()
```

```python
# plotting First and Second periods tweets count per month
# Source: https://stackoverflow.com/questions/43214978/how-to-display-custom-values-on-a-bar-plot Retrieved July 2022
```

```python
#plotting First period tweets count per month

covidMaskP1['created_at'] = pd.Categorical(covidMaskP1['created_at'],
                                   categories=['Jun', 'Jul', 'Aug', 'Sep'],
                                   ordered=True)
counts = covidMaskP1['created_at'].value_counts().rename_axis('Month').reset_index(name='Tweets count')

ax = sns.barplot(x='Month', y='Tweets count', data=counts)
ax.bar_label(ax.containers[0])
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
ax.set_title('First period tweets count per month')
plt.show()
```

```python
#plotting Second period tweets count per month

covidMaskP2['created_at'] = pd.Categorical(covidMaskP2['created_at'],
                                   categories=['Jun', 'Jul', 'Aug', 'Sep'],
                                   ordered=True)
counts = covidMaskP2['created_at'].value_counts().rename_axis('Month').reset_index(name='Tweets count')

ax = sns.barplot(x='Month', y='Tweets count', data=counts)
ax.bar_label(ax.containers[0])
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
ax.set_title('Second period tweets count per month')
plt.show()

```

```python
# plotting First and Second periods  Sentiments Frequency and plotting First and Second periods Sentiments Level (intensity) Frequency
# Source: https://www.geeksforgeeks.org/display-percentage-above-bar-chart-in-matplotlib/ Retrieved July 2022
```

```python
# plotting First period Sentiments Frequency

data = covidMaskP1['Sentiment'].value_counts().rename_axis('Sentiment').reset_index(name='SentimentFrequency')

# compute total sentiments 
total_sent = data.SentimentFrequency.sum()
 
# compute percentage for each sentiment
per = []
for i in range(data.shape[0]):
    cal = (data.SentimentFrequency[i] / total_sent) * 100
    per.append(round(cal,2))
data['Percentage'] = per
 
# display diagram
plt.figure(figsize=(8,8))
colors = ['green','Orange', 'Blue']
diagram = plt.bar(data.Sentiment,data.SentimentFrequency, color = colors)
plt.title('First period Sentiments Frequency')
plt.xlabel("Sentiments")
plt.ylabel("Sentiments Frequency")
plt.xticks(rotation = 45)
i = 0
for p in diagram:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy()
    plt.text(x+width/2,
             y+height*1.01,
             str(data.Percentage[i])+'%',
             ha='center')
    i+=1
plt.show()
```
