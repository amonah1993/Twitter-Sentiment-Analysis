```python
# Getting the CSV files ready for hydrating the tweet IDs
# Source: https://ieee-dataport.org/open-access/coronavirus-covid-19-tweets-dataset Retrieved July 2022

import pandas as pd
#IEEE data port COVID-19 with sentiments datasets
df1=pd.read_csv("corona_tweets_469.csv", header=None)
df2=pd.read_csv("corona_tweets_470.csv", header=None)
df3=pd.read_csv("corona_tweets_105.csv", header=None)
df4=pd.read_csv("corona_tweets_468.csv", header=None)
df5=pd.read_csv("corona_tweets_104.csv", header=None)
df6=pd.read_csv("corona_tweets_103.csv", header=None)

# only keeping the first column, removing the sentiment column and saving in new data frames
df1Edited=df1[0]
df2Edited=df2[0]
df3Edited=df3[0]
df4Edited=df4[0]
df5Edited=df5[0]
df6Edited=df6[0]

#saving the new datasets without sentiment column into a new csv file 
df1Edited.to_csv("ready_corona_tweets_469.csv", index=False, header=None)
df2Edited.to_csv("ready_corona_tweets_470.csv", index=False, header=None)
df3Edited.to_csv("ready_corona_tweets_105.csv", index=False, header=None)
df4Edited.to_csv("ready_corona_tweets_468.csv", index=False, header=None)
df5Edited.to_csv("ready_corona_tweets_468.csv", index=False, header=None)
df6Edited.to_csv("ready_corona_tweets_103.csv", index=False, header=None)
```

```python
# concatenation all orginial IEEE data port COVID-19 with sentiments datasets and rename the columns
dfConcat=pd.concat([df1,df2,df3,df4,df5,df6])
dfConcat.rename(columns={0:'id',1:'Sentiment'}, inplace=True)
```

```python
dfConcat
```

```python
# loading the resulted dataset from the hydration process
dfAll=pd.read_csv("ready_corona_tweets_all.csv")
```

```python
dfAll
```

```python
# merge the hydrated dataset and the orginal IEEE dataset based on id column 
dfMerged = pd.merge(dfAll, dfConcat, on=['id'], how='left')
```

```python
# saving the merged dataset into a new csv file
dfMerged.to_csv("CovidMaskTrainingDatasetRaw.csv",mode='w', encoding='utf-8', index=False, line_terminator='\n')
```
